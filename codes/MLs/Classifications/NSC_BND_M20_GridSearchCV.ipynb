#### 정형화

```python
df_BND_M20_dropna.columns
```

```python
target = df_BND_M20_dropna['SICK_SYM1_3_target']
features = df_BND_M20_dropna[['EDC_ADD_RT', 'ED_RC_TOT_AMT']]
target.shape, features.shape
```

```python
from sklearn.model_selection import train_test_split
features_train,features_test, target_train,target_test = train_test_split(features, target, random_state=111)
features_train.shape,features_test.shape, target_train.shape,target_test.shape
```

#### 모델학습

```python
from sklearn.tree import DecisionTreeClassifier, DecisionTreeRegressor #범주형 Classifier 연속형 Regressor
```

```python
model = DecisionTreeClassifier()
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
hyper_params = {'min_samples_leaf' : [5, 7, 9],'max_depth' : [9, 11],'min_samples_split' : [5, 6, 7] }
```

###### 평가 score : 분류-정확도, 예측-R squre

```python
from sklearn.metrics import f1_score
```

```python
grid_search = GridSearchCV(model, param_grid=hyper_params, cv=3, verbose=1, scoring=f1_score)
```

```python
grid_search.fit(features_train,target_train)
```

```python
grid_search.best_estimator_
```

```python
grid_search.best_score_, grid_search.best_params_
```

```python
best_model = grid_search.best_estimator_
best_model
```

```python
target_test_predict = best_model.predict(features_test)
target_test_predict
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(target_test, target_test_predict))
```
