```python
df_BND_M20 = pd.DataFrame(list)
```

```python
df_BND_M20 = df_BND_M20[['SICK_SYM1','MCARE_SUBJ_CD','TOT_PRSC_DD_CNT','ED_RC_TOT_AMT','EDC_SBA', 'EDC_INSUR_BRDN_AMT']].copy()
```

```python
df_BND_M20[:2]
```

```python
df_BND_M20['SICK_SYM1'].value_counts()
```

```python
def convertSICK_SYM(sick_sym) :
    if len(sick_sym) < 3 :
        return None
    else :
        return sick_sym[:3]
```

```python
convertSICK_SYM('J209'), convertSICK_SYM('f_')
```

```python
df_BND_M20['SICK_SYM1_3'] = df_BND_M20['SICK_SYM1'].apply(convertSICK_SYM)
```

```python
df_BND_M20['SICK_SYM1_3'].value_counts()
```

```python
df_BND_M20_dropna = df_BND_M20.dropna()
df_BND_M20_dropna.isnull().sum()
```

```python
target_list = df_BND_M20_dropna['SICK_SYM1_3'].value_counts().index.to_list()
```

```python
target_list.index('L20')
```

```python
def appyIndexNumber(sick_sym_3):
    indexNumber = target_list.index(sick_sym_3)
    return indexNumber
```

```python
 df_BND_M20_dropna['SICK_SYM1_4'] = df_BND_M20_dropna['SICK_SYM1_3'].apply(appyIndexNumber)
```

```python
 df_BND_M20_dropna['SICK_SYM1_4'].value_counts()
```

```python
df_BND_M20_dropna.info()
```

```python
df_BND_M20_dropna.columns
```

```python
from sklearn.model_selection import train_test_split
target = df_BND_M20_dropna['SICK_SYM1_4']
features = df_BND_M20_dropna[['MCARE_SUBJ_CD', 'TOT_PRSC_DD_CNT', 'ED_RC_TOT_AMT','EDC_SBA', 'EDC_INSUR_BRDN_AMT']]
```

```python
features_train, features_test, target_train, target_test = train_test_split(features, target, random_state=111)
features_train.shape, target_train.shape, features_test.shape, target_test.shape
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(features_train,target_train)
```

```python
target_train_predict = model.predict(features_train)
target_train_predict.shape # target_train.shape 동일
```

```python
target_test_predict = model.predict(features_test)
target_test_predict.shape # target_test.shape 동일
```

```python
from sklearn.metrics import accuracy_score # 정확도
```

```python
accuracy_score(target_train,target_train_predict) # 교내 시험
```

```python
accuracy_score(target_test, target_test_predict)
```

```python
from sklearn.metrics import confusion_matrix, precision_score, recall_score, f1_score, classification_report
```

```python
confusion_matrix(target_train, target_train_predict)
```

```python
precision_score(target_train, target_train_predict, average='micro')
```

```python
recall_score(target_train, target_train_predict, average='micro')
```

```python
f1_score(target_train, target_train_predict, average='micro')
```

```python
print(classification_report(target_train,target_train_predict))
```

```python
print(classification_report(target_test,target_test_predict))
```
